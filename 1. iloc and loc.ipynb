# Importing Dataset using Pandas

```python
from pandas import read_csv
import pandas as pd
dataset=pd.read_csv('Uber Drives 2016.csv')
dataset
```

# iloc
## Selecting Rows

```python
dataset.iloc[0]   # first row of data frame
```

```python
dataset.iloc[1]   # second row of data frame
```

```python
dataset.iloc[-1]  # last row of data frame
```

## Selecting Columns

```python
dataset.iloc[:,0]   # first column of data frame
```

```python
dataset.iloc[:,1]   # second column of data frame
```

```python
dataset.iloc[:,-1]  # last column of data frame
```

```python
dataset.iloc[:,-2]  # second last column of data frame
```

## Multiple row and column selections

```python
dataset.iloc[0:5]   # first five rows of dataframe 
```

```python
dataset.iloc[:, 0:2]   # first two columns of data frame with all rows
```

```python
dataset.iloc[:, 2:-1]   # 2nd to second last column of data frame with all rows
```

```python
dataset.iloc[[0,5,9,24], [0,1]]   # 1st, 6th, 10th, 25th row + 1st and 2nd columns
```

```python
dataset.iloc[0:7, 3:6]   # first 7 rows and 3rd, 4th, 5th columns of data frame
```

# using loc

```python
dataset.loc[:,'START*']   # START* column of data frame with all rows
```

```python
dataset.loc[:,['START*','STOP*']]   # START*, STOP* columns of data frame with all rows
```

```python
dataset.loc[0:5,['START_DATE*','MILES*']]   # first 5 rows and START_DATE*, MILES* columns 
```

# We can also store selected data in other variable

```python
m = dataset.iloc[0:5, 2:5] 
m
```

```python
l = dataset.loc[0:3,['START_DATE*','END_DATE*','CATEGORY*','START*']]
l
```
